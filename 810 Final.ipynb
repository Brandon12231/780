```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/drive')
data_folder = '/content/drive/MyDrive/BA810/Data/'
```

```python
rawdata = pd.read_csv(data_folder+'default_of_credit_card_clients.csv', index_col=0)

rawdata.head()
```

```python
rawdata.info()
rawdata.describe()
```

```python
# change type of "default payment next month", EDUCATION, MARRIAGE, and SEX into category
rawdata['default payment next month'] = rawdata['default payment next month'].astype('category')
rawdata['EDUCATION'] = rawdata['EDUCATION'].astype('category')
rawdata['MARRIAGE'] = rawdata['MARRIAGE'].astype('category')
rawdata['SEX'] = rawdata['SEX'].astype('category')
rawdata.info()
```

```python
from sklearn.model_selection import train_test_split
X = rawdata.drop('default payment next month', axis=1) # separate X ...
y = rawdata['default payment next month'].copy()       # from y
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = .2, random_state=42)
train_X.shape, test_X.shape, train_y.shape, test_y.shape
```

```python
# Create the random forest classifier
clf = RandomForestClassifier(n_estimators=100, class_weight='balanced')
clf.fit(train_X, train_y)
```

```python
from sklearn import metrics

#set up plotting area
plt.figure(0).clf()

# Predict probabilities for the test set
y_prob = clf.predict_proba(test_X)

# Compute the false positive rate, true positive rate, and threshold for each entry in the ROC curve
fpr, tpr, threshold = metrics.roc_curve(test_y, y_prob[:,1])

# Plot the ROC curve
plt.plot(fpr, tpr, 'b-')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```
